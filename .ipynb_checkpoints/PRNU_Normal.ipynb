```python
import os
import glob
import numpy as np
from PIL import Image
from prnu import extract_multiple_aligned, crosscorr_2d

print("âœ… Successfully imported PRNU functions.")

## ---------------------------------------------------------------------------
## PART 1: CREATE AND SAVE FINGERPRINTS FOR EACH CAMERA
## ---------------------------------------------------------------------------

data_directory = r"/Users/shreetsu/prnu-camera-source-detection/data"
fingerprint_directory = os.path.join(os.path.dirname(data_directory), 'fingerprints')
os.makedirs(fingerprint_directory, exist_ok=True)

device_folders = [f for f in os.listdir(data_directory) if os.path.isdir(os.path.join(data_directory, f))]
print(f"Found device folders: {device_folders}")

for device in device_folders:
    print(f"\nProcessing fingerprint images for: {device}...")

    # Get all JPG/JPEG images regardless of case
    image_paths = glob.glob(os.path.join(data_directory, device, 'images', 'fingerprint_set', '*.[jJ][pP][gG]')) + \
                  glob.glob(os.path.join(data_directory, device, 'images', 'fingerprint_set', '*.[jJ][pP][eE][gG]'))

    if not image_paths:
        print(f"âš ï¸  No images found for {device} fingerprint_set. Skipping.")
        continue

    TARGET_WIDTH = 1024
    TARGET_HEIGHT = 768

    # Read images using PIL
    images = []
    for path in image_paths:
        try:
            img = Image.open(path).convert('RGB')  # Ensure 3 channels
            img = img.resize((TARGET_WIDTH, TARGET_HEIGHT), Image.LANCZOS)
            img_np = np.array(img, dtype=np.uint8)
            images.append(img_np)
        except Exception as e:
            print(f"âš ï¸ Could not read image: {path} ({e})")

    if not images:
        print(f"âš ï¸ No valid images for {device}. Skipping.")
        continue

    shapes = set([img.shape for img in images])
    if len(shapes) != 1:
        print(f"âŒ Mismatched shapes in {device}: {shapes}")
        continue

    print(f"Number of images for {device}: {len(images)}")
    for idx, im in enumerate(images):
        print(f"  Image {idx} shape: {im.shape}, dtype: {im.dtype}")
    
    # Extract the camera fingerprint
    fingerprint = extract_multiple_aligned(images, processes=0)

    # Save the fingerprint
    save_path = os.path.join(fingerprint_directory, f"{device}_fingerprint.npy")
    np.save(save_path, fingerprint)
    print(f"âœ… Fingerprint for {device} saved to: {save_path}")


```

```python
## ---------------------------------------------------------------------------
## PART 2: MATCH AN UNKNOWN IMAGE
## ---------------------------------------------------------------------------
print("\n--- Starting Image Matching Test ---")

unknown_image_path = r"/Users/shreetsu/prnu-camera-source-detection/data/iphone15/images/query_set/6A38C3E6-08D8-419C-A3EE-2E09D8CDCF14.jpeg"
print(f"Testing image: {unknown_image_path}")

# Read query image with PIL and resize to match fingerprint size
TARGET_WIDTH = 1024
TARGET_HEIGHT = 768

try:
    unknown_img = Image.open(unknown_image_path).convert('RGB')
    unknown_img = unknown_img.resize((TARGET_WIDTH, TARGET_HEIGHT), Image.LANCZOS)
    unknown_np = np.array(unknown_img, dtype=np.uint8)
except Exception as e:
    raise ValueError(f"Could not read query image: {unknown_image_path} ({e})")

# âœ… Extract noise residual from the unknown image
from prnu.functions import noise_extract_compact, rgb2gray

query_noise = noise_extract_compact((unknown_np, 4, 2))  # levels=4, sigma=2 is standard
query_noise_gray = rgb2gray(query_noise)

# Load all fingerprints
fingerprints = {}
for fp_file in os.listdir(fingerprint_directory):
    if fp_file.endswith('.npy'):
        device_name = fp_file.replace('_fingerprint.npy', '')
        fingerprints[device_name] = np.load(os.path.join(fingerprint_directory, fp_file))

print(f"Loaded {len(fingerprints)} fingerprints for matching.")

# Compare unknown image against each fingerprint
best_match_device = None
highest_correlation = -1

for device, fingerprint in fingerprints.items():
    # Convert fingerprint to grayscale if not already
    if fingerprint.ndim == 3:
        fp_gray = rgb2gray(fingerprint)
    else:
        fp_gray = fingerprint

    # crosscorr_2d returns a map; take max correlation as score
    correlation_map = crosscorr_2d(query_noise_gray, fp_gray)
    correlation_score = float(np.max(correlation_map))

    print(f"  - Correlation with {device}: {correlation_score:.4f}")

    if correlation_score > highest_correlation:
        highest_correlation = correlation_score
        best_match_device = device

print("\n--- Results ---")
if best_match_device is not None:
    print(f"ðŸš€ The best match for the image is: **{best_match_device}** with a score of {highest_correlation:.4f}")
else:
    print("âš ï¸ No match found.")
```

```python
import os
import glob
from PIL import Image
import numpy as np
import sys

# >>>>>> CRITICAL: Import all necessary functions from the PRNU library <<<<<<
# You must ensure the 'prnu' library folder is correctly linked in your environment's path
try:
    from prnu import crosscorr_2d
    from prnu.functions import noise_extract_compact, rgb2gray 
except ImportError as e:
    sys.exit(f"âŒ FATAL ERROR: Cannot import PRNU functions. Please check your setup. Error: {e}")
# >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

# --- CONFIGURATION ---
data_directory = r"/Users/shreetsu/prnu-camera-source-detection/data"
fingerprint_directory = os.path.join(os.path.dirname(data_directory), 'fingerprints')

# Define the folder to test and its known source
QUERY_FOLDER_PATH = r"/Users/shreetsu/prnu-camera-source-detection/data/OnePlus Nord CE4/images/query_set/"
CORRECT_SOURCE = 'OnePlus Nord CE4' 

# Standard resizing parameters
TARGET_WIDTH = 1024
TARGET_HEIGHT = 768

## ---------------------------------------------------------------------------
## PART 1: LOAD FINGERPRINTS
## ---------------------------------------------------------------------------

fingerprints = {}
for fp_file in os.listdir(fingerprint_directory):
    if fp_file.endswith('.npy'):
        device_name = fp_file.replace('_fingerprint.npy', '')
        fingerprints[device_name] = np.load(os.path.join(fingerprint_directory, fp_file))

if not fingerprints:
    sys.exit("âŒ ERROR: No fingerprints loaded. Check the 'fingerprint_directory' path.")

print(f"Loaded {len(fingerprints)} fingerprints for matching: {list(fingerprints.keys())}")


## ---------------------------------------------------------------------------
## PART 2: BATCH MATCHING (CROSS-CORRELATION) AND ACCURACY
## ---------------------------------------------------------------------------
print("\n--- Starting Batch Traditional Correlation (PCE) Test ---")
print(f"Known correct source: {CORRECT_SOURCE}")

query_image_paths = glob.glob(os.path.join(QUERY_FOLDER_PATH, '*.[jJ][pP][gG]')) + \
                    glob.glob(os.path.join(QUERY_FOLDER_PATH, '*.[jJ][pP][eE][gG]'))

total_images = 0
correct_identifications = 0

if not query_image_paths:
    print("âš ï¸ No query images found in the specified folder.")
else:
    for image_path in query_image_paths:
        total_images += 1
        
        # 1. Image and Noise Residual Extraction 
        try:
            unknown_img = Image.open(image_path).convert('RGB')
            unknown_img = unknown_img.resize((TARGET_WIDTH, TARGET_HEIGHT), Image.LANCZOS)
            unknown_np = np.array(unknown_img, dtype=np.uint8)
        except:
            continue
        
        # Extract noise residual and convert to grayscale (Uses the actual PRNU functions)
        query_noise = noise_extract_compact((unknown_np, 4, 2))
        query_noise_gray = rgb2gray(query_noise)

        # 2. Compare against all fingerprints
        best_match_device = None
        highest_correlation = -1.0
        
        for device, fingerprint in fingerprints.items():

            if device == 'Samsung S21 FE':
                continue
            
            # Convert fingerprint to grayscale
            fp_gray = rgb2gray(fingerprint) if fingerprint.ndim == 3 else fingerprint

            # Calculate Cross-Correlation Score (PCE)
            correlation_map = crosscorr_2d(query_noise_gray, fp_gray)
            correlation_score = float(np.max(correlation_map))

            if correlation_score > highest_correlation:
                highest_correlation = correlation_score
                best_match_device = device
        
        # 3. Tally Accuracy
        is_correct = (best_match_device == CORRECT_SOURCE)
        
        if is_correct:
            correct_identifications += 1
            
        print(f"  Processed {os.path.basename(image_path):<30}: Identified as {best_match_device} ({'Correct' if is_correct else 'Incorrect'}). Score: {highest_correlation:.4f}")

## ---------------------------------------------------------------------------
## FINAL ACCURACY REPORT
## ---------------------------------------------------------------------------

print("\n" + "="*80)
print("           FINAL ACCURACY REPORT - TRADITIONAL CORRELATION")
print("="*80)

if total_images > 0:
    accuracy = (correct_identifications / total_images) * 100
    print(f"Total Images Processed: {total_images}")
    print(f"Correct Identifications: {correct_identifications}")
    print(f"Accuracy for {CORRECT_SOURCE} folder: **{accuracy:.2f}%** ðŸŽ¯")
else:
    print("Could not calculate accuracy as zero valid images were processed.")
```
